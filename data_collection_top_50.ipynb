```python
decoded_music_request_text = json.loads(music_request.text)
```

```python
track_items = decoded_music_request_text['tracks']['items']
track_items
```

```python
top_50_data = {
    'track_name':[],
    'release_date':[],
    'artist_name':[],
    'artist_id':[]
}
for track in track_items:
    track_name = track['track']['name']
    top_50_data['track_name'].append(track_name)
    print(track_name)
     
    release_date = track['track']['album']['release_date']
    top_50_data['release_date'].append(release_date)
    print(release_date)
    
    artist = str(track['track']['album']['artists']).split(',')[3]
    artist_name = artist[10:-1]
    top_50_data['artist_name'].append(artist_name)    
    print(artist_name)
    
    artist = str(track['track']['album']['artists']).split(',')[2]
    artist_id = artist[8:-1]
    top_50_data['artist_id'].append(artist_id)
    print(artist_id)
    
    print('----------')
```
