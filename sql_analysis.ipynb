## SQL #1

Question: Create a table that displays who the top performing artists are?

Business Justification: By seeing the artists who have the most songs in the top 50 chart, we can see which artist creates the most top tracks. We will pursue that artist to try to create deals and try to tailor some of the Apple Music marketing with the use of those better performing artists. 

SQL features: Common Table Expression, Join, Group by

```python
%%sql
#CTE to see which artists have the most top songs
WITH artist_top_song_count AS (
    SELECT artist_name, track_name
    FROM top_50_songs 
    JOIN artists 
        ON artists.artist_id = top_50_songs.artist_id
)

SELECT artist_name, COUNT(*) AS num_top_songs
FROM artist_top_song_count
GROUP BY artist_name
ORDER BY num_top_songs DESC;
```

## SQL #1

Reccomendaion:

Based on these results we see that the artist that performs the most top songs is Bad Bunny. As for my reccomendation, I believe that Apple Music should strike some sort of deal with Bad Bunny, for example, have him release his music exclusively on Apple Music for a week before releasing on other platforms. Additionally, if Bad Bunny would not agree to that, Dua Lipa,  who follows Bad Bunny's 5 top songs with 3 top songs, could be another target for a similar promotional plan.

## SQL #2

Question: Create a classification for popular artists, who are artists that have a popularity rating greater than that of industry standard.

Business Justification: We can filter out the artists that have a popularity rating lower than that of the average for the specific genre. Then by seeing which artists have the highest popularity rating we can see which artists have a the majority of people listening to their music. This will indicate which artists are good to pursue in partnerships regardless if they do not have many songs in the current top 50 chart.

SQL Features: Case, Common Table Expression, Join

```python
%%sql
WITH artist_genre_popularity AS (
    SELECT 
        artist_name, 
        genres.genre_name,
        artist_popularity,
        AVG(artist_popularity) OVER(PARTITION BY artists.genre_id) AS avg_artist_popularity_per_genre,

            CASE
                WHEN artists.artist_popularity < 90 THEN 'Mildly Popular'
                WHEN artists.artist_popularity BETWEEN 90 AND 95 THEN 'More Popular'
                ELSE 'Extremely Popular'
            END AS 'popularity_level'

    FROM artists
    JOIN genres
        ON artists.genre_id = genres.genre_id
)

SELECT *
FROM artist_genre_popularity
WHERE artist_popularity > avg_artist_popularity_per_genre
ORDER BY artist_popularity DESC;
```

## SQL #2

Reccomendation: From this table we see once again that Bad Bunny has a high popularity rating that is higher than the genre average. Additionally, we see that there is only one EDM artist on this list. Although the artist is mildly popular compared to the rest of the artist, MEDUZA does have a popularity level higher than the average for EDM. Therefore targeting any of these artists would be a good bet as they all have popularity ratings higher than genre average.

If looking at just the popularity rating, Bad Bunny would be the clear cut pick for a branding deal. 

However, artists like Travis Scott, Drake, and The Weeknd all are categorized as extremely popular and have a popularity rating that is 10 points higher than their respecitve genre average. Therefore, these would be the best artists for Apple Music to pursue to get deals with or create playlist that are more tailored to those listening to these artists.

## SQL #3

Question: Create a query that displays the the top genres and the percentage of each age group that listens to that genre and the average artist popularity for that specific genre.

Business Justification: By showing the most popular genres and their respective age groups, we can choose which age group to market towards and form a target market for Apple Music using that. Additionally, marketing can be ramped up for specific genres in specific age groups while other genres and age groups not so much.

SQL features: View, Join, Group By


```python
%%sql
CREATE OR REPLACE VIEW top_songs_by_artists_with_genre AS
    SELECT artists.artist_name, top_50_songs.track_name, genres.genre_id, genres.genre_name, genres.age_16_19, 
        genres.age_20_24, genres.age_25_34, genres.age_35_44, genres.age_45_54, genres.age_55_64, genres.age_65_plus
    FROM top_50_songs
    JOIN artists
        ON top_50_songs.artist_id = artists.artist_id
    JOIN genres
        ON artists.genre_id = genres.genre_id;

SELECT 
    genre_id, genre_name, age_16_19, 
    age_20_24, age_25_34, age_35_44,
    age_45_54, age_55_64, age_65_plus, 
    COUNT(*) AS num_of_top_songs_in_genre
FROM top_songs_by_artists_with_genre
GROUP BY genre_id
ORDER BY COUNT(*) DESC;

```

## SQL #3

Reccomendation: 

We see by far that pop has the most songs in the top 50 charts. Therefore if resources were to be given towards marketing the most popular genre it would be pop. Additionally, the age groups demonstrate that more than half of those from age 16-34 listen to pop music. Therefore, marketing should be done towards Millenials and Gen Z.
